```python
#Import Libraries

# First let's import the packages we will use in this project
# You can do this all now or as you need them
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure


matplotlib.rcParams['figure.figsize'] = (12,8)

#Read the data

df = pd.read_csv(r'/Users/chintanpatel/Downloads/movies.csv')
```

```python
 #looking for null values any

 
 for col in df.columns:
        print("Col-> ",col)
        print("Col-> ",col)
        print()
        missing_val = np.mean(df[col].isnull())
        print(missing_val)
        
```

```python
# chnage the value of Na or Null to zero

df[["budget","gross"]] = df[["budget","gross"]].fillna(0)


# data types for our columns
df.dtypes
```

```python
# Need to remove 0 from budget column

#change data type of column

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
# Release date and year should be match with the year column

#extracting year using reg in pandas
# Create correct year column
df['yearcoorect'] = df['released'].astype('str').str.extract(r'(\d{4})')   # four digit
df.head()
```

```python
df = df.sort_values(by=['gross'], inplace = False, ascending=False)
```

```python
# work on company

# as it is object lets do some numerization

df_numerize = df  # dont messup with my df

for col_name in df_numerize.columns:
    if(df_numerize[col_name].dtype=='object' or df_numerize[col_name].dtype=='category' ):
        df_numerize[col_name]= df_numerize[col_name].astype('category')
        df_numerize[col_name]= df_numerize[col_name].cat.codes
        
        
df_numerize.head()
```

```python
df.dtypes
```

```python
# unstacking 

df_numerize_matrix = df_numerize.corr()
df_numerize_unstack = df_numerize_matrix.unstack()
df_numerize_unstack
```

```python
#sorting

df_sorted = df_numerize_unstack.sort_values()
df_sorted
```

```python
#high co-relation

high_corr = df_sorted[(df_sorted)>0.5]
high_corr = high_corr[(high_corr)<0.98]
high_corr
```
